## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

bars[0].set_hatch('/')
bars[1].set_hatch('o')

plt.figure(figsize=(6,4))

plt.show()
```

##Load FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

##Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color = '#eb3434')

plt.xticks(bins)

plt.ylabel('Number of players')
plt.xlabel('Skill level')
plt.title('Distribution of FIFA player skills in 2018')

plt.show()
```

##Pie Chart

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

plt.pie([left, right], labels=['Left', 'Right'], colors=["#cdeb34", "#ebc034"], autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')


plt.show
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
explode = (.5,.1,.1,.1,.1)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200'], autopct='%.2f %%', explode=explode)

plt.show
```

```python
plt.figure(figsize=(4,6))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels  = ['FC Barcelona', 'Real Madrid', 'NER']

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color="#4286f4", linewidth=2)
    box.set(facecolor="#c6eb34")
    
plt.title("Football Team Comparison")
plt.ylabel('FIFA Overall Rating')
plt.show()
```
