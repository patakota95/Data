```python
xiaomi_data = brand_model_counts[brand_model_counts['brand_name'] == 'xiaomi']
samsung_data = brand_model_counts[brand_model_counts['brand_name'] == 'samsung']




xiaomi_count = Smartphones['brand_name'].value_counts()['xiaomi']
samsung_count = Smartphones['brand_name'].value_counts()['samsung']


Smartphones = plt.bar(['Xiaomi', 'Samsung'], [xiaomi_count, samsung_count], color=['#42cef5', '#4254f5'])
plt.title('Mobiles Sold by each brand')

for bar in Smartphones:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, yval, ha='center', va='bottom')
    
    
plt.show
```
